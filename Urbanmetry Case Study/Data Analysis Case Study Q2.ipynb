## Q2.1 

#### Choose three buildings and compare amongst them based on the price trend chart and the raw data available in the "Q2 data" worksheet.

```python
#filtering 3 properties

Q2 = Q2.query('NAME == "LILY POINT" or NAME == "ASTRA SUITES" or NAME == "HARBOURVIEW"')
Q2
```

```python
#extracting average price data

Q2A = pd.DataFrame(pd.read_excel("DataCaseStudy.xlsx",sheet_name="Q2"))
Q2A = Q2A.iloc[2:8,2:17].reset_index(drop=True)
Q2A.columns = Q2A.iloc[0]
Q2A = Q2A.query('NAME == "LILY POINT" or NAME == "ASTRA SUITES" or NAME == "HARBOURVIEW"')
Q2A
```

```python
#filtering values between the year 2016 Q1 and 2018 Q4

Q2_1 = Q2[(Q2.QUARTER >= 20161) & (Q2.QUARTER <= 20192)]
Q2_1
```

```python
#finding the mean value of built up price psf

Q2_2 = pd.DataFrame(Q2_1.groupby(['QUARTER','NAME'])['Built up price psf'].mean())
Q2_2 .rename(columns={"Built up price psf":"mean"},inplace=True)
Q2_2
```

```python
#plotting the mean values to compare

fig, ax =plt.subplots(figsize=(10,5))
sns.set_style('whitegrid')
sns.lineplot(data=Q2_2, x="QUARTER", y="mean",hue="NAME")
```

**HARBOURVIEW** had the highest rate since the first quarter of 2016 **(2016 Q1)** and till the last quarter of 2018 **(2018 Q4)**. **ASTRA SUITES** had the average rate around 1000psf since the first quarter of 2016 and **hiked** suddenly by the middle of 2018. **LILY POINT** had the lowest rate of all time which even **decreased by the year 2018**.

## Q2.2
#### Plot another price trend that shows the median price psf across time for these three buildings.

```python
#finding the median of the built up price psf 

Q2_3 = pd.DataFrame(Q2_1.groupby(['QUARTER','NAME'])['Built up price psf'].median())
Q2_3.rename(columns={"Built up price psf":"median"},inplace=True)

Q2_3
```

```python
#plotting both mean and median values of built up price of similar properties to observe

fig, ax =plt.subplots(1,2,figsize=(29,10))

sns.lineplot(data=Q2_3, x="QUARTER", y="median",hue="NAME",ax=ax[0])
sns.lineplot(data=Q2_2, x="QUARTER", y="mean",hue="NAME",ax=ax[1] )

fig.show()
plt.grid()
```

Based on the graph of mean() and median(), in mid-2018, **Harbourview and Astra Suites** experienced a psf rate **hike**, while **Lily Point** saw a **decline** in its rate, indicating changes in property values for these locations.

## Q2.3
#### Plot another price trend that shows the median valuation price across time for these three buildings.

```python
#finding the valuation price of the same properties

Q2_4 = pd.DataFrame(Q2_1.groupby(["QUARTER","NAME"])['Valuation price'].median())
Q2_4.reset_index(inplace=True)
Q2_4.rename(columns={"Valuation price":"Val median"},inplace=True)
Q2_4
```

```python
#plotting the valuation price to compare

fig, ax =plt.subplots(figsize=(10,5))
sns.lineplot(data=Q2_4, x="QUARTER", y="Val median",hue="NAME")
```

Both the **LILY POINT & ASTRA SUITES** has low valuation price while **HARBOURVIEW** has the highest comparatively.
